## Scatter Plots

```python
fig, axs = plt.subplots(2, 3, figsize=(20,10))
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample size vs. Median', ax=axs[0,0])
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample size vs. Unemployment rate', ax=axs[0,1])
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full time vs. Median', ax=axs[0,2])
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Share women vs. Unemployment rate', ax=axs[1,0])
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', ax=axs[1][1])
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', ax=axs[1,2])
```

I am not sure we can answer required questions using visualizations above. Let's create two more scatter plots.

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Total vs. Median', ax=axs[0])
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Share women vs. Median', ax=axs[1])
```

### Plots QA

* Q: Do students in more popular majors make more money?
* A: There is no correlation between total amount of students in a major (neither Sample size) and their median salary. Of course, it depends of how do we detect 'popular' majors.


* Q: Do students that majored in subjects that were majority female make more money?
* A: We can see overly small corelation between share women coefficient and median salary. However, students that majored in subjects with majority female make a bit less money.


* Q: Is there any link between the number of full-time employees and median salary?
* A: No, there is no link between of the number of full-time employees and median salary.

## Histograms

```python
fig, axs = plt.subplots(3, 3, figsize=(20,15))
axs[-1][-1].axis('off')
recent_grads['Sample_size'].plot(bins=12, range=(0,3000), grid=True, kind='hist', title='Sample size', rot=30, ax=axs[0,0])
recent_grads['Median'].plot(bins=12, range=(20000,80000), grid=True, kind='hist', title='Median', rot=30, ax=axs[0,1])
recent_grads['Employed'].plot(bins=20, range=(0,200000), grid=True, kind='hist', title='Employed', rot=30, ax=axs[0,2])
recent_grads['Full_time'].plot(bins=20, range=(0,200000), grid=True, kind='hist', title='Full time', rot=30, ax=axs[1,0])
recent_grads['ShareWomen'].plot(bins=10, range=(0,1), grid=True, kind='hist', title='Share women', rot=30, ax=axs[1,1])
recent_grads['Unemployment_rate'].plot(bins=5, range=(0,0.25), grid=True, kind='hist', title='Unemployment rate', rot=30, ax=axs[1,2])
recent_grads['Men'].plot(bins=20, grid=True, kind='hist', title='Men', rot=30, ax=axs[2,0])
recent_grads['Women'].plot(bins=20, grid=True, kind='hist', title='Women', rot=30, ax=axs[2,1])
```

### Histograms QA

* Q: How many majors are predominantly male? Predominantly female?
* A: Looking at the Share women histogram we can tell there are ~50% of majors are predominantly male and other ~50% are predominantly female.

```python
print('Predominantly male: {} ({:.0f}%)'.format(
    sum(recent_grads['ShareWomen'] <= 0.5), 
    100 * (sum(recent_grads['ShareWomen'] <= 0.5) / len(recent_grads))))
print('Predominantly female: {} ({:.0f}%)'.format(
    sum(recent_grads['ShareWomen'] > 0.5), 
    100 * (sum(recent_grads['ShareWomen'] > 0.5) / len(recent_grads))))
```
